```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
#We see here that new customers tend to spend more than regular or loyal customers. The more Loyal you get the less you spend on average
```

```python
#We now want to target customers based on spending. First step- group!
```

```python
ords_prods_merge['mean_orders'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_orders'] > 10, 'spender_flag'] = 'Low Spender'
ords_prods_merge.loc[ords_prods_merge['mean_orders'] <= 10,'spender_flag'] = 'High Spender'
```

```python
ords_prods_merge['spender_flag'].value_counts(dropna=False)
```

```python
ords_prods_merge['median_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)

```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.loc[ords_prods_merge['median_orders'] > 20, 'frequency_flag'] = 'Non-Frequent Customer'
ords_prods_merge.loc[(ords_prods_merge['median_orders']<=20) & (ords_prods_merge['median_orders'] > 10), 'frequency_flag'] = 'Regular Customer'
ords_prods_merge.loc[ords_prods_merge['median_orders'] <= 10,'frequency_flag'] = 'Frequent Customer'
```

```python
ords_prods_merge['frequency_flag'].value_counts(dropna=False)
```
